# **Time Sereis Analysis with python**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('/content/multiTimeline.csv',skiprows=1)
df.head(10)
```

```python
df.columns=['week','Python','Java','C','Sql']
df.head()
```

```python
df.dtypes
```

```python
df.month=pd.to_datetime(df.week)
df.set_index('week',inplace=True)
df.head()
```

```python
df['Python']=df['Python'].replace('<1',0)
df['Sql']=df['Sql'].replace('<1',0)
df.head()
```

```python
df.tail()
```

# **lets check the trend of covid 19 Search on google Overtime**

```python
python=df['Python']
python.rolling(15).mean().plot(figsize=(20,10),linewidth=2,fontsize=20)
plt.xlabel('year',fontsize=20)
```

```python
java=df['Java']
java.rolling(15).mean().plot(figsize=(20,10),linewidth=2,fontsize=20,color='red')
plt.xlabel('year',fontsize=20)
```

```python
c=df['C']
c.rolling(15).mean().plot(figsize=(20,10),linewidth=2,fontsize=20,color="green")
plt.xlabel("year",fontsize=20)
```

```python
Sql=df['Sql']
Sql.rolling(15).mean().plot(figsize=(20,10),linewidth=2,fontsize=10,color='orange')
plt.xlabel('year',fontsize=20)
```

```python
df.plot(figsize=(20,10),linewidth=3,fontsize=20)
plt.xlabel('year',fontsize=20)
```

```python
python.diff().plot(figsize=(20,10),linewidth=3,fontsize=20)
plt.xlabel('year',fontsize=20)
```
